```python
import pandas as pd
from pathlib import Path
from semevalpolar.utils import get_project_root
import os
```

```python
data_path = Path(os.path.join(get_project_root(), "data", "test_phase", "subtask1", "train", "eng.csv"))
dev_df = pd.read_csv(data_path)
dev_df = dev_df.sample(frac=1, random_state=42)
```

```python
dev_df_polar = dev_df[dev_df["polarization"] == 1]
dev_df_not_polar = dev_df[dev_df["polarization"] == 0]
dev_df_not_polar = dev_df_not_polar.sample(n=len(dev_df_polar), random_state=42)
```

```python
balanced = pd.concat([dev_df_polar, dev_df_not_polar], ignore_index=True)
balanced.to_csv(f"{get_project_root()}/src/semevalpolar/finetuning/instruct/data/train/balanced.csv", index=False)
```
