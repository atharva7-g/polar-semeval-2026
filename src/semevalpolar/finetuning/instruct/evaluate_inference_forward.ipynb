```python
import json
import os
import re
from pathlib import Path

import pandas as pd

from semevalpolar.finetuning.instruct.inference_forward import (
    generate_predictions_jsonl_forward_pass,
)
from typing import List

from semevalpolar.utils import get_project_root
```

```python
data_path = Path(
    os.path.join(get_project_root(), "data", "test_phase", "subtask1", "dev", "eng.csv")
)
dev_df = pd.read_csv(data_path)
dev_df = dev_df.sample(frac=1, random_state=42)
```

```python
records = dev_df["text"]
gold = dev_df["polarization"]
```

```python
generate_predictions_jsonl_forward_pass(
    inputs=records, output_path=f"{get_project_root()}/df_predictions.jsonl"
)
```

```python
data_path = Path(f"{get_project_root()}/df_predictions.jsonl")
with data_path.open("r", encoding="utf-8") as f:
    records = f.readlines()
    records_dicts = [json.loads(s) for s in records]
```

```python
labels = [record["predicted_label"] for record in records_dicts]
```

```python
comparison = pd.DataFrame(
    {
        "Predicted": labels,
        "Ground Truth": gold,
    }
)
```
