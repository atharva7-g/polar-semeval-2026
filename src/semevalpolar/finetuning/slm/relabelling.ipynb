```python
from semevalpolar.llm.data_utils import read_dataset, split_dataframe
import os

from semevalpolar.utils import get_project_root

df = read_dataset(os.path.join(get_project_root(), "data", "relabelling", "eng.csv"))
```

```python
data_df = df[df["polarization_split"].notnull()]
```

```python
clean_data_df = data_df[data_df["polarization_split"] != 1]
train_df, val_df, test_df = split_dataframe(clean_data_df)
```

```python
clean_data_df
```

```python
train_df = train_df.rename(columns={"polarization": "label"})
val_df = val_df.rename(columns={"polarization": "label"})
test_df = test_df.rename(columns={"polarization": "label"})

for df in [train_df, val_df, test_df]:
    df["label"] = df["label"].astype(int)
```

```python
from datasets import DatasetDict, Dataset

dataset = DatasetDict({
    "train": Dataset.from_pandas(train_df),
    "validation": Dataset.from_pandas(val_df),
    "test": Dataset.from_pandas(test_df),
})

dataset = dataset.remove_columns("__index_level_0__")
```
