```python
import pandas as pd

from semevalpolar.utils import get_project_root
import os
from semevalpolar.llm.data_utils import read_dataset, create_submission
from semevalpolar.finetuning.slm.inference import run_inference_on_df
```

```python
project_root = get_project_root()
```

```python
df = read_dataset(os.path.join(get_project_root(), 'data', 'dev_phase', 'subtask1', 'dev', 'eng.csv'))
df.head()
```

```python
model_dir = os.path.join(
    project_root,
    "predictions",
    "finetuning",
    "checkpoints",
)

prediction = run_inference_on_df(
    df=df,
    text_column="text",
    model_dir=model_dir,
    batch_size=16,
)
```

```python
type(prediction)
```

```python
submission = create_submission(df, prediction)
```

```python
submission.to_csv(os.path.join(project_root, 'predictions', 'subtask_1', 'pred_eng.csv'))
```
