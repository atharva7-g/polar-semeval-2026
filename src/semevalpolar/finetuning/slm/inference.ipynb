```python
import os

import numpy as np
import pandas as pd

from semevalpolar.finetuning.slm.inference import run_inference_on_df
from semevalpolar.llm.data_utils import (
    read_dataset,
    create_submission,
    create_comparison_df,
)
from semevalpolar.utils import get_project_root
```

```python
df = read_dataset(
    os.path.join(get_project_root(), "data", "dev_phase", "subtask1", "dev", "eng.csv")
)
df = df.sample(len(df), random_state=86)
```

# Testing Finetuned model

```python
model_dir = os.path.join(
    get_project_root(),
    "predictions",
    "finetuning",
    "final_model",
)
```

```python
prediction = run_inference_on_df(
    df=df,
    text_column="text",
    model_dir=model_dir,
    batch_size=16,
)
```

```python
binary_preds = np.where(np.array(prediction) == 2, 1, 0)
```

```python
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    f1_score,
    recall_score,
)

comparison = create_comparison_df(binary_preds, df["polarization"], df["text"])
cm = confusion_matrix(
    comparison["Ground Truth"], comparison["Predicted"], labels=[0, 1]
)
precision = precision_score(
    comparison["Ground Truth"], comparison["Predicted"], labels=[0, 1]
)
recall = recall_score(
    comparison["Ground Truth"], comparison["Predicted"], labels=[0, 1]
)

print(cm)
print((precision, recall))
```
